```python
import numpy as np
import pandas as pd 
from sklearn.linear_model import LogisticRegression
from joblib import dump

from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('train_dataset2.csv')
df
```

```python
df["CODE_GENDER"] =  df["CODE_GENDER"].replace(['F','M'],[0,1])
df["FLAG_OWN_CAR"] = df["FLAG_OWN_CAR"].replace(["Y","N"],[1,0])
df["FLAG_OWN_REALTY"] = df["FLAG_OWN_REALTY"].replace(["Y","N"],[1,0])
```

```python
df.drop('ID', axis=1, inplace=True)
```

```python
names = df.keys()
print(names)
```

```python
types = df.dtypes
print(types)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for i in range(len(types)):
    if types[i] == 'object':
        # Chuyển đổi cột thành chuỗi nếu nó không phải kiểu chuỗi
        df[names[i]] = df[names[i]].astype(str)
        
        le.fit(df[names[i]])
        df[names[i]] = le.transform(df[names[i]])
```

```python
data = np.array(df.values)
data
```

```python
X = data[:,:-1]
y = data[:,-1]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

print('Train:')
print(X_train.shape)
print(y_train.shape)
print('Test:')
print(X_test.shape)
print(y_test.shape)
```

```python
from sklearn import tree
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train,y_train)


# Lưu mô hình vào một file
dump(clf, 'dt.joblib')

clf_pred = clf.predict(X_test)

print('DT')
print('Accuracy Score is {:.5}'.format(accuracy_score(y_test, clf_pred)))
print(classification_report(y_test, clf_pred))
print(confusion_matrix(y_test, clf_pred))
```

```python
# Initialize the Logistic Regression model
logistic_model = LogisticRegression()

# Train the model
logistic_model.fit(X_train, y_train)

dump(logistic_model, 'logistic.joblib')

logistic_pred = logistic_model.predict(X_test)

# Evaluate the model
print('Accuracy Score is {:.5}'.format(accuracy_score(y_test, logistic_pred)))
print(classification_report(y_test, logistic_pred))
print(confusion_matrix(y_test, logistic_pred))
```
